```python
#simple neural networf(NN)
#classifying MNIST digits(0-9)

#import libraries
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Flatten
from tensorflow.keras.datasets import mnist
```

```python
# 2. load the MNIST dataset(images of digits)

(x_train,y_train),(x_test,y_test) = mnist.load_data()

#images are0-255 - normalize them 0-1
x_train=x_train/255.0
x_test=x_test/255.0
```

```python
# 3.build a SIMPLE nural network moel

model=Sequential([
    #flatten: converts 28x28 image - 784 iput vector
    Flatten(input_shape=(28,28)),
            
    #hidden layer: 128 neurons + reulu activation
    Dense(128,activation="relu"),
    
    # output layer: 10 neurons (for digit 0-9) + softmax
    Dense(10,activation="softmax")

])
    
'''
layers:
faltten layer
converts 2d image - 1d vector
NN always takes 1D input'''
```

```python
# 4. compile the model

model.compile(
    optimizer="adam",  # how NN learns
    loss="sparse_categorical_crossentropy",  # Multi-class loss
    metrics=["accuracy"]  # what to measure
)
'''
step 4 - compile
we choose:
optimizer = adam(most used)
loss=cross entropy(for multiclss)
metric

'''
```

```python
# 5. train the model

model.fit(x_train,y_train,epochs=5)
'''
step 5 - train
NN adjusts its weights using backpropogation
'''
```

```python
# 6. evalute the model

test_loss,test_acc=model.evaluate(x_test,y_test)

print("Test accuracy:",test_acc)
```
