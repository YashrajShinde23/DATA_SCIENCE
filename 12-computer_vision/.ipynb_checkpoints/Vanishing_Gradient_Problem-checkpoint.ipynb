VANISHING GRADIENT PROBLEM:
1.the vanishing gradient problem occurs when the gradients (the values used to update the weights during traing) become extremely small as they move backword through the metwork during backpropogation
2.when gradient shrink too much:
-wights stop updating
-learning bexcomes extremely slow
-deep nural networks fail to train properly

SOFTMAX
Advantages:
1.converts outputs to probablities
2.sum of probablities=1
3.suitable for choosing exactly one corrct class
Disadvantages
1.senstive to outliers
2.used only in output layer,not hidden layers
3.used in:Multi-class classificationn(e.g,MNIST digits)


```python
import numpy as np
import matplotlib.pyplot as plt


# softmax function
def softmax(x):
    exp_x=np.exp(x - np.max(x))
    return exp_x / np.sum(exp_x)


#example class scores
scores=np.array([1,2,3,4])

y= softmax(scores)

#plot (bar graph)
plt.bar(range(len(scores)),y)
plt.title("Softmax Activaton (Probablity Distribution)")
plt.xlabel("Class")
plt.ylabel("Probablity")
plt.grid(True)
plt.show()


#advantages
print("\nSoftmax Advantages:")
print("1. converts raw numbers into probablites.")
print("2. sum of outputs = 1 - interpretable.")

#diasadvantages
print("\nSoftmax Disadvantages:")
print("1. very sensitive to large inputs(exp grows fast).")
print("2. not used in hiddden layers, only output layer.")
print("3. can cause nuemrical overflow.")
    
```
