```python
#dataset preparteion
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,SimpleRNN
```

```python
#generate sequntial data
#create sequence loke [1,,2,3]->4
X=[]
y=[]

for i in range(1,100):
    X.append([i,i+1,1+2])
    y.append(i+3)
    
X=np.array(X)
y=np.array(y)
```

```python
#simple ANN 
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

ann_model = Sequential([
    Dense(10, activation="relu", input_shape=(3,)),
    Dense(1)
])

ann_model.compile(
    optimizer="adam",
    loss="mse"
)

ann_model.fit(X, y, epochs=50, verbose=0)
```

```python
#test ANN
pred_ann=ann_model.predict(np.array([[10,11,12]]),verbose=0)
print("ANN prediction for [10,11,12]:",pred_ann)
#it is going to show 11 wont understand thr sequence
#NN problem (explain)
#Ann does not know:
#which number came first
#which came last
#it only sees 3 unrelated inpiuts
```

```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import SimpleRNN, Dense
from tensorflow.keras.optimizers import Adam

# data preparation
X = []
y = []

for i in range(1, 100):
    X.append([i, i+1, i+2])
    y.append(i+3)

X = np.array(X, dtype=np.float32)
y = np.array(y, dtype=np.float32)

# normalize
X = X / 100.0
y = y / 100.0

# reshape for RNN
X_rnn = X.reshape((X.shape[0], X.shape[1], 1))

# RNN model
rnn_model = Sequential([
    SimpleRNN(16, activation="tanh", input_shape=(3,1)),
    Dense(1)
])

optimizer = Adam(lr=0.001, clipnorm=1.0)

rnn_model.compile(
    optimizer=optimizer,
    loss="mse"
)

rnn_model.fit(X_rnn, y, epochs=200, verbose=0)

# test data
test = np.array([[10,11,12]], dtype=np.float32) / 100.0
test = test.reshape((1,3,1))

pred = rnn_model.predict(test, verbose=0)

# rescale output
print("RNN prediction:", float(pred[0][0] * 100))
```

```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import SimpleRNN, Dense
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.layers import LSTM

```

```python
#lstm model
# data preparation
X = []
y = []

for i in range(1, 100):
    X.append([i, i+1, i+2])
    y.append(i+3)

X = np.array(X, dtype=np.float32)
y = np.array(y, dtype=np.float32)

# normalize
X = X / 100.0
y = y / 100.0

# reshape for RNN
x_lstm=X.reshape((X.shape[0],X.shape[1],1))
```

```python
#lstm model
model=Sequential([
    LSTM(8,input_shape=(3,1)),
    Dense(1)
])

optimizer=Adam(lr=0.0005)

model.compile(optimizer=optimizer,loss="mse")

model.fit(x_lstm,y,epochs=300,verbose=0)

#test
test=np.array([[10,11,12]],dtype=np.float32)/100.0
test=test.reshape((1,3,1))

pred=model.predict(test,verbose=0)
print("LSTM Prediction:",float(pred[0][0]*100))
```

```python
#Lstm is not meant to learn exact arithmetic like +1
#lstm leanrs patterns statistically,not mathematically
#lstm prediction:16.65
#lstm is too powerful for this simple numeric task
#it tries to fit a dynamic function,not a fixed formula
#if simple model exists,use a simple model
#deep models are for complex,noisy,non-linear sequnces

'''
|model       |suitability|
|------------|-----------|
|linear regression
'''
```
