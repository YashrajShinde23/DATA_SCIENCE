loss function:
1.a loss function tells the neural network how wrong its predictions are
2.it is a numerical value that measures the difference between:
-predicted outputs
-actual/true output
-the goal of training a nural network is to minimize the loss

why do we need a loss function
1.without a loss function the model would have no idea
2.it is performing well
3.how much error occurs

how loss function works during training
1.model predicts sometging
2.loss function calculates error
3.backpropogatin adjusts weights
4.loss decreases - model improves
5.this process repeats for many epochs

```python
import numpy as np

#step 1: true values (actual)
y_true = np.array([3,5,2,7])

#step 2 : predicted values
y_pred = np.array([2,4,3,10])


#step3 : calculates MSE manually
errors = y_true - y_pred
squared_errors = errors*2
mse = np.mean(squared_errors)


print("True values: ",y_true)
print("Predicted values ",y_pred)
print("Errors: ",errors)
print("Squared Errors: ",squared_errors)
print("\nMean Squared Error (MSE) =",mse)
```
