```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# create simple dataset: y= 2x+1
x=np.linspace(-1,1,100)
y=2*x+1
```

```python
#simple 1-layer model: y_pred -w*x+b

def train_model(lr,epochs=50):
    w,b=0.0,0.0 # intail weights
    loss_values=[] #for storing loss
    
    for epochs in range(epochs):
        y_pred=w*x+b
        loss=np.mean((y-y_pred**2)) #MSE loss
        loss_values.append(loss)
        
        #gradients
        dw=np.mean(2*(y_pred-y)*x)
        db=np.mean(2*(y_pred-y))
        
        
        #update weights
        w-=lr*dw
        b-=lr*db
        
    return loss_values
```

```python
#try differnt leanring rates

lrs=[0.0001,0.01,1.0]
results={}

for lr in lrs:
    results[lr]=train_model(lr)
```

```python
#plot results
plt.figure(figsize=(10,6))

for lr in lrs:
    plt.plot(results[lr],label=f"LR={lr}")
             
plt.xlabel("epochs")
plt.ylabel("loss")
plt.title("learning rate experiment: loss vs epochs")
plt.legend()
plt.grid(True)
plt.show
             
```
