```python
import numpy as np

#step 1: true labels(0 or 1)

y_true = np.array([1,0,1,0])
```

```python
#step 2: predicted probablities

y_pred = np.array([0.9,0.2,0.8,0.4])
```

```python
#step 3: compute BCE manually

bce=-np.mean(y_true*np.log(y_pred) +
            (1-y_true) * np.log(1-y_pred))

print("True labels:       ",y_true)
print("Predicted probablities:",y_pred)
print("\nBiary cross Entropy =",bce)
```
