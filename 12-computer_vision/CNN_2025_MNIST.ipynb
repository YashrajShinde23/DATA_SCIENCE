```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D,MaxPooling2D
from tensorflow.keras.layers import Flatten,Dense
from tensorflow.keras.utils import to_categorical

```

```python
#step 2: load mnist dataset
#mnist contains 28x28 grayscale images ofdigits(0-9)
#it is already split into training and testing data
(X_train,y_train),(X_test,y_test)=mnist.load_data()

#print shapes to understand tyhe data
print(X_train.shape)
print(X_test.shape)
```

```python
#step 3: data preprocessing
#pixel values are bestween 0 and 255
#dividing by 255 scales them bestween 0 and 1
#this helps the network train faster
X_train=X_train/255.0
X_test=X_test/255.0
```

```python
#cnn expects 40 input
#(number_of_scales,height,width,chaannels)
#since mnist is grayscale channels=1
X_train=X_train.reshape(-1,28,28,1)
X_test=X_test.reshape(-1,28,28,1)
#cnn does not work with flattened input diecrtly
```

```python
#convert labels(0-9) into oneh-hot encodeing
y_train=tp_categorical(y_train,10)
y_test=to_categorical(y_test,10)
```

```python
#step 4: build cnn model

#sequential means layers are stacked one after another
model=Sequential()
```

```python
#convolutional layer
#32 filters -> network learns 32 different types
#(3x3) is the filter size
#relu removes negative values
model.add(Conv2D(filters=32,
                kernal_size=(3,3),
                activation="relu",
                input_shape=(28,28,1)))
```

```python
#maxpooling layers
#reduces image size by taking vlaue in 2x2 window
#keeps important features and reduces computatin
model.add(MaxPolling2D(pool_size=(2,2)))
```

```python
#flatten layer
#needed before passing to Dnese layer
model.add(Flatten())
```

```python
#fully connected (Dense) hidden layer
model.add(Dense(128,activation="relu"))
```

```python
#step 5: compile the model
model.compile(optimizer="adam",
             loss="categorical_crossentropy",
             metrics=["accuracy"])
```

```python
#step 6: train the cnn
#epochs -> how many times model sees entire daatset
#batch_size -> number of times images preocessed at once
history=model.fit(X_train,y_train,
                 epochs=5,
                 batch_size=64,
                 validation_data=(X_test,y_test))
```

```python
# evalute the model

#test the model on unseen data
test_loss,test_accuracy=model.evaluate(X_test,y_test)

print("test accuracy", test_accuracy)
```
