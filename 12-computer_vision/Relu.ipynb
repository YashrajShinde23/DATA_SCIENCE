```python
import numpy as np
import matplotlib.pyplot as plt

#input values
x = np.linspace(-10,10,400)

#Relu function
def relu(x):
    return np.maximum(0,x)

y=relu(x)


#plot
plt.plot(x,y)
plt.title("Relu Activation Function")
plt.xlabel("Input")
plt.ylabel("Output")
plt.grid(True)
plt.show()


#advantages
print("\nRelu Advantages:")
print("1. very simple and fast to compute.")
print("2. does not saturate for positive values.")
print("3. helps deep networks learn faster (no vanishing gradient for x>0).")

#disadvantes
print("\nRelu Disadvantages:")
print("1. for negative x, output becomes 0 - neuron 'dies'.")
print("2. not zero-centered.")
print("3. gradient is 0 for x < 0 - learning stops there.")
```
