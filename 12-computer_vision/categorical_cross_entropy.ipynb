```python
import numpy as np

#step 1: true label in one-hot format
#example : 3 classes - class 1 is format

y_true=np.array([0,1,0])

#step 2: predicted probablities from softmax
#(must sum to 1)

y_pred=np.array([0.1,0.7,0.2])

#step 3: compute categorical cross entropy

cce=-np.sum(y_true * np.log(y_pred))

print("True label (one-hot):  ",y_true)
print("Predicted probablities:  ",y_pred)
print("\nCategorical cross Entopy =",cce)
```
