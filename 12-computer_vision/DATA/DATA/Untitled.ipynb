```python
# put this all in one cell

import cv2

#connects to your computers default camera
cap=cv2.VideoCapture(0)


#automatically grab width and height from video feed
#(returns float which we need to convert to integer for later on!)
width=int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height=int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

while True:
    
    #capture frame by frame
    ret,frame=cap.read()
    #our operation on the frame come here
    gray=cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
    #display the resulting frame
    cv2.imshow("frame",gray)
    
    '''
    read the key pressed
    if it is the letter "q"
    them break the loop (stop recording/processing)
    '''
    
    if cv2.waitKey(1) & 0xFF == ord("q"):
        break
        
#when everything done, release the capture and destroy the windows
cap.release()
cv2.destroyAIWindows()

```

```python
import cv2
cap=cv2.VideoCapture(0)
#cheak if camera opened correctly
if not cap.isOpened():
    print("error: could not open camera,")
    exit()

width=int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height=int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))


# use a valid codec (XVID is okay)
writer=cv2.VideoWriter(
    "c:/Data-Science/12-computer_vision/DATA/DATA/student_capture.mp4",
    cv2.VideoWriter_fourcc(*"XVID"),
    23,
    (width,height)
)
while True:
    ret,frame=cap.read()
    if not ret:
        print("failed to grab frame")
        break
        
    writer.write(frame)
    cv2.imshow("frame",frame)
    '''
    read the key pressed
    if it is the letter "q"
    then break the loop (stop recording)
    '''
    if cv2.waitKey(1) & 0xFF == ord("q"):
        break
        
    #
    cap.release()
    writer.release()
    cv2.destroyAllWindows()


```
