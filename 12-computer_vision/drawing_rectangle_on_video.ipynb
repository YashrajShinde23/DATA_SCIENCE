```python
import cv2
cap=cv2.VideoCapture(0)


#autmaticallt grab width and height from video feed
#(returns float which we need to convert to integer for later on!)
width=int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height=int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

#we are using // here beacuse in python // allows for int classical divison,
#beacuse we cant pass a float to th ecv2.rectangle function

#coordinates for rectangle,center pointrectangle
x=width//4
y=height//4

#width and height adding 1/4 of width and height from the centre
w=width//4
h=height//4

while True:
    #capture frame by frame
    ret,frame=cap.read()
    
    #draw a rectangle  stream
    
    cv2.rectangle(frame,(x,y),(x+w, y+h),color=(0,0,255),thickness=4)
    
    #display the resulting frame
    cv2.imshow("frame",frame)
    
    #this command lets us quit the "q" button on a keyboard
    #simply pressing x on the window wont work
    if cv2.waitKey(1) & 0xff == ord("q"):
        break
        
#when everything is done release the capture
cap.release()
cv2.destroyAllWindows()
```

```python
import cv2

#
#global variables to store points
#

pt1=(0,0)
pt2=(0,0)
topLeft_clicked=False
botRight_clicked=False


#mouse callback function : detects clicksand sets rectangle points

def draw_rectangle(event,x,y,flags,param):
    
    
    global pt1,pt2,topLeft_clicked,botRight_clicked
    
    #detect left mouse button click
    if event == cv2.EVENT_LBUTTONDOWN:
        
        #case1:both corners were sleected erlier reset everything
        #this allows draing a new rectangle on the next two clicks
        if topLeft_clicked and botRight_clicked:
            topLeft_clicked=False
            botRight_clicked=False
            pt1=(0,0)
            pt2=(0,0)
            
    #case 2: foirst click set top left corner
    if not topLeft_clicked:
        pt1=(x,y)    #save first corner 
        topLeft_clicked=True  # mark first coner as slected
        
    #case 3:second click set botoom right corner
    elif not botRight_clicked:
        pt2=(x,y) # save second corner
        botRight_clicked=True  # mark second as selected
        
        
        
#webcam setup
    
cap=cv2.VideoCapture(0)
    
#create window where output will appear
cv2.namedWindow("Test")
    
    
# attach mpouse callback function to the window
cv2.setMouseCallback("Test",draw_rectangle)
    
    
    # main loop:display video draw shapes when clicking
    
while True:
    ret,frame = cap.read()# read frame from wbcam
        
    #if first point is seleted draw a red dot
        
    if topLeft_clicked:
            cv2.circle(frame,center=pt1,radius=5,color=(0,0,255),thickness=-1)
        
    #if both points slected draw rectangle 
    if topLeft_clicked and botRight_clicked:
        cv2.rectangle(frame,pt1,pt2,(0,0,255),2)
            
    #show updated frame
    cv2.imshow("Test",frame)
        
    #quit loopp when "q" key is pressed
    if cv2.waitKey(1) & 0xff == ord("q"):
        break
            
#clean up
cap.release()
cv2.destroyAllWindows()
        
```
