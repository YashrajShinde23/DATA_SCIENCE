Tanh(Hyperbolic Tangent)
Advantages:
1.output between -1 and 1
2.zero-centered - smoother trainng
3.better than sigmoid for many hidden layers
Disadvantages:
1.still suffers from vanishing gradient
2.saturates at large positive/negatve values
3.used in :RNNs,hidden layers(sometimes)


```python
import numpy as np 
import matplotlib.pyplot as plt

#input values
x = np.linspace(-10,10,400)

# Tanh Function
def tanh_function(x):
    return np.tanh(x)

y=tanh_function(x)


#plot
plt.plot(x,y)
plt.title("Tanh Activation Functon")
plt.xlabel("Input")
plt.ylabel("Output")
plt.grid(True)
plt.show()


#advantages
print("\nTanh Advantages:")
print("1. zero-centered output - better traing than sigmoid.")
print("2. output between -1 and 1")

#disadvantages
print("\nTanh Disadvantages:")
print("1. still suffers from vanishing gradients.")
print("2. saturates at large +ve x values.")
```
