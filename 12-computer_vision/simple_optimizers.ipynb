```python
import numpy as np
import tensorflow as tf
```

```python
# simple data: y = 2x - 1
X=np.array([[1.0],[2.0],[3.0],[4.0]],dtype=np.float32)
Y=np.array([[1.0],[3.0],[5.0],[7.0]],dtype=np.float32)
```

```python
def train_and_show(optimizer,name,epochs=100):
    #create a very simple model: y_pred =w*x+b
    #it has 1 dense neuron with 1 input feature
    model = tf.keras.Sequential([
        tf.keras.layers.Dense(1,input_shape=[1])
    ])
    
    
    #compile the model:
    # - 'optimizer' is passes from outside (SGD,Adam,etc)
    # - 'mse' mens mean squared error loss function
    model.compile(optimizer=optimizer,loss='mse')
    
    #train the model on (X,Y)
    #epochs = how many times the model sees the enire datset
    #verbose=0 - do not print training messages
    model.fit(X,Y,epochs=epochs,verbose=0)
    
    #predict output for x=5
    #note: predict() expects 2d input - shape must be (batch_size,features)
    pred=model.predict(np.array([[5.0]]),verbose=0) #return shape(1,1)
    
    
    #extract learned parameters from the dense layer
    #get_weights() returns [weights_matrix,bias_vector]
    #for this single-nuron model,shapes are:
    # w:(1,1),b:(1,)
    w,b=model.layers[0].get_weights()
    
    
    #display model results
    print(f"\n{name}")
    print("Predicted for x=5 ->",pred.flatten()[0])
    print("Learned weight(w) =", float(w.flatten()[0]))
    print("Learned bias(b) ",float(b[0]))
          
          
```

```python
# SGD
train_and_show(tf.keras.optimizers.SGD(lr=0.1),"SGD (lr=0.1)")
```

```python
#(c) RMSProp
train_and_show(tf.keras.optimizers.RMSprop(lr=0.01),"RMSProp (lr=0.01)")
```

```python
#(D) Adam
train_and_show(tf.keras.optimizers.Adam(lr=0.01),"Adam (lr=0.01)")
```
